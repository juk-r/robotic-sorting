```python
import simpy
import simpy.resources.store
import simpy.core
import random

from brains import DirectionBrain
from maps import DirectionMap, OneWayMap
from mail_factories import RandomAlwaysReadyMail
from robot import Robot
from structures import Mail, Map, RobotType, Position, Direction
from import_json import import_map, import_json
```

```python
env = simpy.Environment()
mail_factory = RandomAlwaysReadyMail(env, range(1, 10))
map_ = import_map(env, import_json("data\\map1.json"), mail_factory)[0]
direction_map = DirectionMap.generate_shortest(OneWayMap.generate_random(map_))
brain = DirectionBrain(env, direction_map)
robotType = RobotType(1, 1, 1, 1)
robot = Robot(env, robotType, brain, Position(0,0), Direction.down, 1)
```

```python
last_time = 0
while env.now < 100:
    env.step()
    if env.now != last_time:
        print(env.now, robot.position, robot.direction, robot.mail, sep='\t')
        last_time = env.now
```
